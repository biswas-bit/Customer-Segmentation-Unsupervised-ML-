```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../datasets/raw/data_with_labels(k-means).csv")
df.head()
```

```python
stars_avg = df[df["label"]==0]['Spending Score (1-100)'].mean()
total_avg = df['Spending Score (1-100)'].mean()
print(f"Cluster 1 spends {stars_avg/total_avg:.1f}x more than the average customer.")
print()
stars_avg_inc = df[df["label"]==0]['Annual Income (k$)'].mean()
total_avg_inc =df['Annual Income (k$)'].mean()
print()
print(f"Cluster 1 Annual Income is {stars_avg_inc/total_avg_inc:.1f}x more than the average customer.")
print()
print(df[df["label"]==0]['Age'].describe())
# we will map this customers as a Vips
```
