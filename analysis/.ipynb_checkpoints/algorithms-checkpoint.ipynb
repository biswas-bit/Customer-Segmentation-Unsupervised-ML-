```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn 
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("../datasets/processed/preprocessed_data.csv")
```

```python
k_model = KMeans(
    n_clusters=8,
     init='k-means++', 
     n_init='auto', 
     max_iter=300,
     tol=0.0001,
      verbose=0, 
      random_state=None,
      copy_x=True, 
      algorithm='lloyd'
)
clusters = k_model.fit_predict(df)
```

```python
df["cluster_label"] = clusters
```
